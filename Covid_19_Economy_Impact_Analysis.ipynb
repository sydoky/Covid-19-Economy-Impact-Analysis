```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

data=pd.read_csv("/content/transformed_data.csv")
data2=pd.read_csv("/content/raw_data.csv")
```

```python
data["COUNTRY"].value_counts().mode()
```

# **Create a new dataset by combining the necessary columns from both the dataset:**

```python
data.head(2)
```

```python
data2.head(2)
```

```python
code=data["CODE"].unique().tolist()
country=data["COUNTRY"].unique().tolist()
hdi=[]
tc=[]
td=[]
sti=[]
population=data["POP"].unique().tolist()

gdp=[]

for i in country:
  hdi.append((data.loc[data["COUNTRY"]==i,"HDI"]).sum()/294)
  tc.append((data2.loc[data2["location"]==i,"total_cases"]).sum())
  td.append((data2.loc[data2["location"]==i,"total_deaths"]).sum())
  sti.append((data.loc[data["COUNTRY"]==i,"STI"]).sum()/294)
  population.append((data2.loc[data2["location"]==i,"population"]).sum()/294)


aggregated_data=pd.DataFrame(list(zip(code,country,hdi,tc,td,sti,population)),
                             columns=["Country Code","Country","HDI","Total Cases","Total Deaths","Stringency Index","Population"])


aggregated_data.head()
                               


#We created this cell and loop because we're creating a new dataset
#where there is no duplicated values. Only dataset represetns one country once. 
#In our dataset above we have multiple same rows.
```

```python
#we combined two dataset from data and data2
aggregated_data
```

# **Now we start analyzing**

# **Let's have a look at all the countries with the highest number of covid cases:**


```python
figure=px.scatter(data,y="Total Cases",x="Country",size="Total Cases",title="Countries with Highest Covid Cases")

figure.show()
```

# **The total number of death among the countries with the highest number of covid cases:**

```python
figure=px.bar(data,y="Total Deaths",x="Country",title="Countries with Highest Deaths")
figure.show()
```

# **Let's compare the total number of cases and total number deaths in all these countries:**

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=data["Country"],y=data["Total Cases"],name="Total Cases",marker_color="indianred"))

fig.add_trace(go.Bar(x=data["Country"],y=data["Total Deaths"],name="Total Deaths",marker_color="blue"))

fig.update_layout(barmode="group",xaxis_tickangle=-45)
fig.show()
```

# **Let's have a look at the percentage of total deaths and total cases among all the countries with the highest number of Covid-19 cases.**

```python
cases=data["Total Cases"].sum()

decreased=data["Total Deaths"].sum()

labels=["Total Cases","Total Deaths"]
values=[cases,decreased]

fig=px.pie(data,values=values,names=labels,title="Percentage of Total Cases and Deaths")
fig.show()
```

```python
cases=data["Total Cases"].sum()

decreased=data["Total Deaths"].sum()

labels=["Total Cases","Total Deaths"]
values=[cases,decreased]

fig=px.pie(data,values=values,names=labels,title="Percentage of Total Cases and Deaths",hole=.5)
fig.show()
```

# **Let's compare the GDP per capita before Covid-19 and during Covid-19 to have a look the impact of Covid-19 on the GDP per capita:**

```python
fig=go.Figure()
fig.add_traces(go.Bar(
    x=data["Country"],
    y=data["GDP Before Covid"],
    name="GDP Per Capita Before Covid-19",
    marker_color="indianred"
))

fig.add_traces(go.Bar(
    x=data["Country"],
    y=data["GDP During Covid"],
    name="GDP Per Capita During Covid-19",
    marker_color="lightsalmon"
))

fig.update_layout(barmode="group",xaxis_tickangle=-45)
```
